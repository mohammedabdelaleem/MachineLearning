```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('market_data.csv')
df
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X
```

```python
# Normal Equation : (X.T * X) inverse * X.T * y

## insert ones column for the intercept
X = np.insert(X, 0, 1, axis=1)
X
```

```python
X_Transpose = np.transpose(X) # X.T
X_Transpose
```

```python
X_Transpose_dot_X = X_Transpose @ X
```

```python
temp1 = np.linalg.inv(X_Transpose_dot_X)
```

```python
temp2 = X_Transpose @ y
```

```python
thetas = temp1 @ temp2 
```

```python
for i in range(len(thetas)):
    print(f"thetas[{i}] = {thetas[i]}\n")
```

```python
def pred_value(thetas, inputs):
    res = 0
    for i in range(len(thetas)-1):
        res = res + (thetas[i+1]*inputs[i])
    
    return (thetas[0] + res )
```

```python

inputs =[ 84.72, 19.20, 48.96 ]
print(pred_value(thetas, inputs))
```
