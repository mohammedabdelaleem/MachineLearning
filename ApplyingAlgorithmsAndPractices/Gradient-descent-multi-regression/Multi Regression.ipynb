```python
df_dummies = pd.get_dummies(df_startups["State"], prefix="State")
df_dummies
```

```python
df_startups = pd.concat([df_startups,df_dummies] ,axis=1 )
df_startups
```

```python
df_startups = df_startups.drop(columns=['State', 'State_California'], axis=1)
```

```python
df_startups
```

```python
df_startups.corr()
```

```python
x = df_startups.drop(columns='Profit', axis=1)
y = df_startups['Profit']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=35)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
mlr = LinearRegression()
model = mlr.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
df = pd.DataFrame({"Y_Test":y_test, "Y_Pred":y_pred, "Diff between Y_Pred, Y_Test":abs(y_pred- y_test)})
df
```

```python
from sklearn.metrics import mean_absolute_error
MAE =mean_absolute_error(y_test, y_pred)
MAE
```

```python
RMAE = np.sqrt(MAE)
RMAE
```
